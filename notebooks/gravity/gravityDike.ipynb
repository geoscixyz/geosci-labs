# Gravity anomaly from a dike

## Introduction

This notebook demonstrates the gravity anomaly generated by a 2D dipping dike that is infinite along one horizontal direction and buried in the subsurface.

Three plots are produced:

(1) Data measured along a profile on the surface across the dike

(2) Cross-sectional diagram of the dike model

(3) Plan view of data measured over a grid on the surface

## Setup

The following parameters are adjustable.

$\Delta \rho$: Density contrast defined as the density of the dike minus the density of the host rock (in $g/cm^3$)

$z_1$: Depth to the top of the dike (in $m$)

$z_2$: Depth to the bottom of the dike (in $m$)

$b$: Horizontal width of the dike (in $m$)

$\beta$: Dipping angle of the dike from vertical (in $degree$)

$Step$: Spacing between measurements on the surface (in $m$)


The vertical component of the gravity anomaly at a point on the surface can be calculated as

$\Delta g_z = 2 G \Delta \rho [ z_2 (\theta_2-\theta_4) - z1 (\theta_1-\theta_3) + \mathbf{sin}\beta \mathbf{cos}\beta \{x (\theta_2-\theta_1) - (x-b) (\theta_4-\theta_3)\} + \mathbf{cos}^2\beta \{x \mathbf{ln}(r_2/r_1) - (x-b) \mathbf{ln}(r_4/r_3) \} ]$ in mGal,

where $G = 6.674 \times 10^{-8} cm^3 \cdot g^{-1} \cdot s^{-2}$ is the gravititional constant and $\theta$'s are in radian.
